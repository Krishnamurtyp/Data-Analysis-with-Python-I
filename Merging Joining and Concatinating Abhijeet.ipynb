 # ConcatenationÂ¶
Concatenation basically glues together DataFrames. Keep in mind that dimensions should match along the axis you are concatenating on. You can use pd.concat and pass in a list of DataFrames to concatenate together:

```python
# we have got these three dataframes now we can combine them also notice that they do not have any common row.
# we can combine this by pandas "pd.concat" method
```

```python
pd.concat([df1,df2,df3]) # see here that it has combined that with rows that is because
#by default the argument of the pd.concat method is axis = 0 which correspond to rows
```

```python
# let us try to combine it with columns
```

```python
pd.concat([df1,df2,df3],axis =1)
```

# Merging
The merge function allows you to merge DataFrames together using a similar logic as merging SQL Tables together.

```python
# suppose we have following datasets with us 
```

```python
df11 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df22 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
df11
```

```python
df22
```

```python
# notice that we have a common 'key' column in both the dataframes left and right
# now we can try merging both the dataframes on key let's see how we can do it in python method is pd.merge
```

```python
pd.merge() # when we shift tab this we see there are four major arguments left right how and on 
# left - dataframe you want to keep on the left hand side and right - dataframe you want to keep on the right hand side 
# how is the merge condition as in inner etc and on mean on which column
```

```python
pd.merge(df11,df22,how='inner', on = 'key') # we can see the results 
```

# Joining
Joining is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame.

```python
# joins here follows the same concept as it follows in the sql 

# let us see some of the examples
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
# these are the two tables let us perform join on them 
# the only difference between the merge and join here is the keys that you use to join on will be an index rather than a column 
```

```python
left.join(right)  # by default the join is "left"
```

```python
left.join(right,how = "outer" )
```

```python
left.join(right, how ='inner')
```
